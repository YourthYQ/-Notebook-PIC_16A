### `DataFrame.describe()` 
This method provides a quick overview of the statistical summary of numerical columns by default, but it can also be applied to object-type columns.

Here's a summary of what describe() typically returns for numerical columns:

`count` - The number of non-null entries. <br>
`mean` - The mean of the entries. <br>
`std` - The standard deviation of the entries. <br>
`min` - The minimum value. <br>
`25%` - (first quartile): The value below which 25% of the data fall. <br>
`50%` - (median): The middle value, dividing the data into two halves. <br>
`75%` - (third quartile): The value below which 75% of the data fall. <br>
`max` - The maximum value.

## Boolean Indexing
While it's good to know how to refer to rows by index, this is not the most useful way to filter data frames. Boolean indexing instead allows us to filter the rows of a data set based on one or more conditions. Boolean indexing in data frames is very similar to Boolean indexing in numpy arrays.

```python
penguins
```

```python
penguins['Culmen Length (mm)'] < 40 # this comparison happens elementwise
# it then returns a series of [indexs, comparison_results]
```

```python
mask = penguins['Culmen Length (mm)'] < 40
penguins[mask]
```

```python
mask = penguins['Island'] == 'Torgersen'
mask
```

```python
penguins[mask].head()
```

```python
mask1 = penguins['Culmen Length (mm)'] < 40
mask2 = penguins['Island'] == 'Torgersen'

penguins[mask1 & mask2].head()
```

```python
penguins[mask1 | mask2].head()
```


Boolean indexing is by far the most useful form of filtering, and should usually be preferred in most practical contexts. It is especially powerful when combined with functions that operate on columns, as we'll see shortly.

## Creating and Modifying Columns
The easiest way to add a new column is using the syntax df["column name"].



```python
print(penguins.columns)
penguins.head()
```

```python
cols = ["Species", "Region", "Island", "Culmen Length (mm)", "Culmen Depth (mm)"]
penguins = penguins[cols]
penguins.head()
# what did this do? only take the columns in col from penguins, and save it as penguins
# "Culmen Depth (mm)" is not in the penguins
```

```python
# makes a new column, similar to making new dictionary key-value pair
penguins["Culmen Length (cm)"] = penguins["Culmen Length (mm)"]/10
```

```python
print(penguins.columns)
penguins.head()
```

```python
penguins["Anvers"] = (penguins["Region"] == "Anvers")
penguins.head()
```

```python
# And we are iterating over species, let's say 'Adelie'
species = "Adelie Penguin (Pygoscelis adeliae)"

# This creates a boolean Series [True, False, True, False]
mask = penguins["Species"] == species

# Now when we apply this mask to the DataFrame, we get only the rows where the mask is True
species_data = penguins[mask]

species_data
```

```python
species_data["Culmen Length (mm)"]
```

```python
 # rearrange columns' order
cols = ["Species", "Island", "Region", "Anvers", "Culmen Length (mm)", "Culmen Length (cm)"]
penguins = penguins[cols] # rearrange column order
penguins.head()
```

## Vertorized text processing for columns in pandas

If you remember your operations from `numpy`, you are actually good to go on operations involving numerical columns. As we saw above, you can essentially create `pandas` columns as `np.array`s.

`pandas` also enables flexible, vectorized processing of text data. The required methods are similar to what you would use for simple strings, with the additional wrinkle that these methods are hiding behind a special `str` attribute. Let's look at some examples, with an eventual goal of cleaning up the very long species names of the penguins.

```python
penguins["Species"] # pandas series object
```

```python
s = "harlin"
print(s)
s.upper()
```

```python
penguins["Species"].str.upper() 
```

```python
s = "hello world!"
print(s)
s.split() # str.split() will split the string(as SPACE by default) and store in an array
```

```python
penguins["Species"].str.split() # call str.split() on every row of penguins["Species"]
```

```python
penguins["Species"].str.split().str.get(0) # pandas series
```

```python
penguins["Species"] = penguins["Species"].str.split().str.get(0) # change species names to something shorter
penguins.head()
```

## Externally Setting Columns

It is also possible to add columns without referencing any objects in the data frame. This is not usually recommended, but may be helpful on certain occasions. 

There are two approaches. 

The first is to add a column using a list or similar collection of length equal to the number of rows in the data frame.

`len()` returns the number of rows in the DataFrame, when the object is a pandas DataFrame.

`len(penguins)` would return the total number of rows in the penguins DataFrame.

`range()` generates a sequence of numbers. It is commonly used for looping a specific number of times in for loops or to create a list of numbers.

```python
s = "abcde"

for i in range(len(s)): # this range(s) means a sequence of integers from 0 to len(s).
    print(s[i])
```

```python
print(len(penguins))
penguins["counter"] = range(len(penguins)) # this didn't reference any existing columns in penguins
penguins.head()
```

```python
x = penguins.loc[[4, 1, 0]]
x["counter"] = range(len(x))
x
```

```python
penguins["counter"] = range(3) # new column has to have the same length as dataframe
penguins.head()
```

The second approach is to use a single value, such as a string or int. In this case, all rows will have the same value in the corresponding column.

```python
penguins["cute"] = "cute" # same value for every row of new "cute" column.
penguins.head()
```
