## Purpose of a Mesh Grid
A mesh grid is used to create a grid of points over a specified domain in two dimensions. <br>
This grid allows for the evaluation of functions of two variables at numerous points across the domain. <br>
It's particularly useful for:
* Visualizing surfaces or contours in two and three dimensions.
* Performing numerical integration and differentiation over a specific area.
* Simulating physical phenomena that vary across a plane.

## How It Works
Given two one-dimensional arrays x and y, which represent the range of values for two variables (let's say X and Y), <br>
a mesh grid creates two two-dimensional arrays:
* The first 2D array (X) repeats the x array across its rows for each value of y.
* The second 2D array (Y) repeats the y array across its columns for each value of x.

This process effectively creates a grid of (X, Y) coordinate pairs, covering the rectangular domain defined by the x and y arrays.



```python
x = np.linspace(0, 1, 5)  # 1D array from 0 to 1 with 5 points
y = np.linspace(0, 1, 3)  # 1D array from 0 to 1 with 3 points

print(x)
print(y)
```

```python
X, Y = np.meshgrid(x, y)

# create a rectangular grid out of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing. 
# When you provide 1D arrays x and y to np.meshgrid(x, y), it returns two 2D arrays X and Y.

print("X array:")
print(X)
print("\nY array:")
print(Y)
```

### **`alpha`** keyword
When comparing distributions of several data sets, it's often useful to use *transparency* via the keyword **`alpha`**. 

**`alpha = 1`** generates fully opaque( ) data markings, 

**`alpha = 0`** is fully transparent (and therefore invisible). 

p.s. you can set the **alpha** by yourself through [0,1] to clarify the transparency you want.

# 2d Histograms

Let's thinkbout 2d version of histograms.

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.scatter(x, y)
```

```python
fig, ax = plt.subplots(1)
h = ax.hist2d(x, y, bins=50, cmap = "Blues")
plt.colorbar(h[3], label = "counts in bin") # h[3]: annoying syntax thing to memorize or look up
```

While there's often no problem with square grids, hexagonal grids often look a bit nicer. Note that the relevant keyword for hexagonal grids is `gridsize` rather than `bins`. 

```python
fig, ax = plt.subplots(1)
h = ax.hexbin(x, y, gridsize = 50, cmap = "Blues")
plt.colorbar(h, label = "counts in bin") 
```
